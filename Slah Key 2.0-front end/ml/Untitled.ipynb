```python
import pandas as pd
```

```python
data = pd.read_csv('Dataset.csv')
del data['Unnamed: 0']
del data['BMI_tags']
del data['Label']
data
```

```python
import matplotlib.pyplot as plt

# Histogram
plt.hist(data['calories_to_maintain_weight'], bins=10, width=100)
plt.xlabel('Calories to Maintain Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Calories to Maintain Weight')
plt.show()


data = data[(data['age'] >= 14) & (data['age'] <= 60)]

# Box plot
#plt.boxplot(data['calories_to_maintain_weight'])
#plt.ylabel('Calories to Maintain Weight')
#plt.title('Box Plot of Calories to Maintain Weight')
#plt.show()
```

```python
data1 = data.copy()
data1['calory_to_balance'] = 0
```

```python
data1['calory_to_balance'] = data1['calories_to_maintain_weight'] * 21.7 / data1['BMI']
data1
```

```python
data1.describe()
data1['needed_calory'] = data1['calories_to_maintain_weight'] - data1['calory_to_balance']
data1.describe()
data2=data1.copy()
```
